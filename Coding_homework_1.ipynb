```python
url1='https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
pandas_data1 = pd.read_csv(url1)

url2 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
pandas_data2 = pd.read_csv(url2)
```

```python
#Convert DataFrames into list variables
data1 = pandas_data1.values.tolist()
data2 = pandas_data2.values.tolist()
```

```python
#Flatten the nested lists to make a simple 1D list
data1 = np.squeeze(data1)
data2 = np.squeeze(data2)
```

```python
#Print the number of elements in data1
print(len(data1))
```

```python
#Print the 3rd, 4th, 5th, and 6th elements of data2
print(data2[2:6])
```

```python
#Print the second-to-last and last elements of data1
print(data1[-2:])
```

```python
#Print the type of data2
print(type(data2))
```

```python
#Plot data1 histogram
plt.hist(data1)
```

```python
#add axes labels to data1 histogram
plt.hist(data1)
plt.xlabel('SST Resting Membrane Potential (mV)')
plt.ylabel('Counts')
```

```python
#plot histogram of data2 with axes labeled
plt.hist(data2)
plt.xlabel('Pvalb Resting Membrane Potential (mV)')
plt.ylabel('Counts')
```

The histogram for SST interneurons is shifted slightly toward less negative (more positive) values compared to the PV interneurons. This indicates that SST cells have a higher resting membrane potential, while Pvalb cells are more hyperpolarized at rest. Overall, the distributions suggest that PV cells are generally more negative at rest, whereas SST cells tend to be slightly depolarized.
